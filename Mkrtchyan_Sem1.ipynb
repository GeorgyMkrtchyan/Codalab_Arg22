## Part 2. RuArg-2022: Dialogue 2022 shared task on Argument Mining for the Russian Language

We are given a text (a comment from social media) regarding one or more of the following topics: vaccination, quarantine, masks. 

We have formulated three claims regarding the COVID-19 pandemic (and anti-epidemic measures in general):

```
1. «Вакцинация полезна для общества» (Vaccination is beneficial for society.)
2. «Введение и соблюдение карантина полезно для общества» (The introduction and observance of quarantine is beneficial for society.)
3. «Ношение масок полезно для общества» (Wearing masks is beneficial for society.)
```

Each sentence was annotated by stance and premise for all three claims. Thus, each sentence has six labels.

The following classes (labels) were used:

```
"For" (2)
"Against" (0)
"Other” (for stance, this class merges classes “neutral”, “unclear” or “for and against”) / “no argument” (for a premise) (1)
"Irrelevant" (for this claim) (-1)
```

 Participants are required to automatically annotate each test sentence by stance and premises for each claim (vaccination, quarantine, masks) separately – in total, **six** labels must be assigned to the sentence. Labels belong to a set of four classes (see above).

### Example

```
“This is exactly why everyone should wear masks, but the main channels broadcast that masks are not needed and useless for healthy people.”
```

Let us annotate the text with 6 labels:

| masks stance | masks premise | vaccines stance | vaccines premise | quarantine stance | quarantine premise | 
| :---: | :---: | :---: | :---: | :---: | :---: |
| 2 | 0 | -1 |  -1|  -1| -1 |

````python
import pandas as pd
````

````python
train_df = pd.read_csv("train.tsv", sep="\t")
val_df = pd.read_csv("val_empty.tsv", sep="\t")
test_df = pd.read_csv("test-no_labels.tsv", sep="\t")
````

````python
train_df.head()
````

````python
def categorizer(data, class_name):
    # Select required columns
    data = data[['text_id', 'text', f'{class_name}_stance', f'{class_name}_argument']]

    # Set your model output as categorical and save in new label col
    data['stance_label'] = pd.Categorical(data[f'{class_name}_stance'])
    data['argument_label'] = pd.Categorical(data[f'{class_name}_argument'])

    # Transform your output to numeric
    data[f'{class_name}_stance'] = data['stance_label'].cat.codes
    data[f'{class_name}_argument'] = data['argument_label'].cat.codes
    return data
````

````python
from tqdm import tqdm
````

````python
label = '__class__'
for class_name in tqdm(["quarantine", "vaccines",  "masks"]):
    tmp_data = categorizer(train_df, class_name=class_name)
    train_data = tmp_data[['text', f'{class_name}_stance', f'{class_name}_argument']]
    for case in ['stance', 'argument']:
        train_df[f'{class_name}_{case}'] = label + train_data[f'{class_name}_{case}'].astype(str)
````
